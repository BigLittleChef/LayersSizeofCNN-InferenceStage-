```python
data_name_list=['','_BA_MAX','_BA_Opp','_PA']
methodName=['Fedavg','MWU','MWU_WO','Median','Cronus','mine']
lineName=['Fedavg','MWU','MWU_WO','Median','Cronus','RMCA']
methodFile=[]

path='E:\老王组内成员研究\鹏飞\Matlab作图\Matlab作图\SHVN noniid'
for i in range(len(methodName)):
    #print(methodName[i])
    methodFile.append(methodName[i]+data_name_list[3]+'_acc'+'.npy')
    #print(methodName[i]+'_acc')
print(methodFile)
```
